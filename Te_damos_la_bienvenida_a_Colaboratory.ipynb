```
class Cliente:
    def __init__(self, nombre, apellido, cedula, gmail, vip, direccion, telefono, celular, edad):
        self.nombre = nombre
        self.apellido = apellido
        # Usamos los setters para validar desde el nacimiento del objeto
        self.cedula = cedula
        self.edad = edad

        # Atributos directos
        self.gmail = gmail
        self.vip = vip  # Booleano (True/False)
        self.direccion = direccion
        self.telefono = telefono
        self.celular = celular

    # --- Encapsulamiento de Cédula ---
    @property
    def cedula(self):
        return self._cedula

    @cedula.setter
    def cedula(self, valor):
        # Convertimos a string para contar dígitos y verificamos que sea numérico
        valor_str = str(valor)
        if len(valor_str) == 10 and valor_str.isdigit():
            self._cedula = valor_str
        else:
            raise ValueError(f"Error: La cédula '{valor}' debe tener exactamente 10 dígitos numéricos.")

    # --- Encapsulamiento de Edad ---
    @property
    def edad(self):
        return self._edad

    @edad.setter
    def edad(self, valor):
        if valor >= 18:
            self._edad = valor
        else:
            raise ValueError(f"Error: El cliente debe ser mayor de 18 años (Edad ingresada: {valor}).")

    def __str__(self):
        status_vip = "SÍ" if self.vip else "NO"
        return (f"Cliente: {self.nombre} {self.apellido}\n"
                f"Cédula:  {self.cedula}\n"
                f"Edad:    {self.edad}\n"
                f"Email:   {self.gmail}\n"
                f"VIP:     {status_vip}\n"
                f"Celular: {self.celular}")

# --- Pruebas de funcionamiento ---

try:
    print("--- Intentando crear un cliente válido ---")
    cliente1 = Cliente("Ana", "García", "1712345678", "ana@gmail.com", True, "Av. Siempre Viva 123", "02234567", "0998765432", 25)
    print(cliente1)

    print("\n--- Intentando crear un cliente menor de edad ---")
    cliente2 = Cliente("Luis", "Pérez", "0912345678", "luis@gmail.com", False, "Calle Falsa 123", "04212345", "0987654321", 15)
except ValueError as e:
    print(e)

try:
    print("\n--- Intentando asignar una cédula incorrecta ---")
    cliente1.cedula = "12345"  # Solo 5 dígitos
except ValueError as e:
    print(e)
```
